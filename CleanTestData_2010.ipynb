```python
import glob
path = '/Users/taniajogesh/Documents/CCNPL/TestData/2010_txt/*.txt' 
files = glob.glob(path)
textlist = []
for i in xrange( 0, len(files)):
    textlist.append(open(files[i], 'r').read())
filenames = glob.glob(path)
filenames = [w.replace('/Users/taniajogesh/Documents/CCNPL/TestData/2010_txt/', '') for w in filenames]
names = [w.replace('.txt', '') for w in filenames]
```

```python
import re
urlList = []
r = re.compile("https?:\/\/*")
for i in xrange( 0, len(textlist)):
    urlList.append([x for x in textlist[i].split("\n") if r.match(x)])
```

```python
urlNames = zip(names,urlList)
```

```python
#urlNamesTupple[5]
```

```python
cleanurl =[]
for name, urllist in urlNames:
    for url in urllist:
        cleanurl.append((name,url))
```

```python
cleanurl[300]
```

```python
AllUrls=[]
for name,url in cleanurl:
    new_url = url.replace(' ','')
    AllUrls.append((name,new_url))
```

```python
from newspaper import Article

def html_to_text( urls ):
    a = Article(urls) 
    a.download()
    a.parse()
    #a.nlp()
    text = a.text
    title = a.title
    authors = a.authors
    date = a.publish_date
    source = a.source_url
    #keywords = a.keywords
    data =(text,title,authors,date, source)
    return(data)
```

```python
import pandas as pd
testURLs = pd.DataFrame(AllUrls)
testURLs.columns = ["month_yr","url"]
testURLs.head()
```

```python
URLcount2010=testURLs.groupby('month_yr').agg('count')
URLcount2010.to_csv("URLcount2010")
```

```python
import os
import pickle

if os.path.exists("Test_data_pickle_2010.p"):
    print("Test data already downloaded")
    Testdata = pickle.load(open("Test_data_pickle_2010.p", "rb"))
else:
    Testdata =[]
    Narticles = testURLs["url"].size
    print "Extracting data from online news weblinks...\n"
    for i in xrange(0, Narticles):
        try:
            Testdata.append((html_to_text(testURLs["url"][i]),i))
            if( (i+1)%100 == 0 ):
                print "Downloading %d of %d\n" % ( i+1, Narticles )
        except:
            pass
    pickle.dump(Testdata, open("Test_data_pickle_2010.p", "wb"))    
```
